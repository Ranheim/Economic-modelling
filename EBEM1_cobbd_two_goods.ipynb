```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (11, 5)  #set default figure size
import numpy as np
import sympy as sym
from sympy import init_printing, latex
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from numpy import random
import pygame
```

```python
def b(e,p1,p2,g2): #budjsettbetingelsen (kurven)
    return e/p1-(p2/p1)*g2
    # t er timer med fritid

def u(a,g1,g2): # utility function
    return g1**(a)*g2**(1-a)

def i(a,g2,c): # indifference curves for constant utility c
    #print (g2, c**(1/a) * (1/g2))
    if a > 0:
        return (c**(1/a)) * (g2**(((a-1)/a)))
    else:
        return 0
    
def max_cd_with_constraint(a,p1,p2,e): #good1, good2, parameter a, price 1 price 2, endowment, cobb douglas
    g1_level = e*a/p1
    g2_level = e*(1-a)/p2
    return g1_level, g2_level

def indirect_utility(a,good_vector):
    return good_vector[0]**(a)*good_vector[1]**(1-a)

a = 0.5
p1 = 12
p2 = 22
e = 200

good_vector = max_cd_with_constraint(a,p1,p2,e)
x1 = round(good_vector[0],2)
format(x1, '.2f')
x2 = round(good_vector[1],2)
acquired_utility = indirect_utility(a,good_vector)

print(good_vector, acquired_utility)

t1 = np.arange(0, e/p2+1 ,0.2)
#t2 = np.arange(good_vector[1]-5, good_vector[1]+5 ,1)

fig = plt.figure(1,[(e/p1 + 4)/2,(e/p2 + 4)/2],facecolor='grey')
fig.suptitle("The consumers choice over two goods, cobb-douglas utility with parameter " + str(a) , fontsize = 30)
#plt.subplot(211)
plt.ylabel("Good 1, p = " + str(p1), fontsize=16)
plt.xlabel("Good 2, p = " + str(p2), fontsize=16)
plt.xlim([0,(e/p2 + 3)])
plt.ylim([0,(e/p1 + 3)])
plt.plot(t1, b(e,p1,p2,t1), 'k')
plt.plot(t1, i(a,t1,acquired_utility), 'k')
plt.plot([x2,x2],[0,x1], 'r--', [0,x2],[x1,x1], 'r--') #plots points [1x,2x] [1y,2y]
#plt.text(0,x1+0.3, 'Kvantum gode 1 = %f' % x1, color='b', fontsize = 12)
plt.text(0+0.3,x1+0.3, "Quantity good 1 =  " + str(x1), color ='b', fontsize = 12)
plt.text(x2+0.3,0+0.3, "Quantity good 2 =  " + str(x2), color ='b', fontsize = 12)
plt.text(e/p2, 1, "The slope of the budget curve is -p2/p1, or " + str(round(-p2/p1,2)), color ='b', fontsize = 12)
plt.text(x2 + 1, x1 + 1, "Achieved utility is " + str(round(acquired_utility,2)), color ='b', fontsize = 12)
```
